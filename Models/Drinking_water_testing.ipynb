```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.load_model("Model_drinking.json")
```

```python
def predicting(X):
    targets = ["No drinking", "Drinking"]
    y_pred = model.predict(X)

    result = [targets[class_id] for class_id in y_pred]

    return result
```
